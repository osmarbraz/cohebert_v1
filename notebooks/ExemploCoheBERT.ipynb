# 2 Exemplo de Uso

## Declara e instancia o Coerência BERT 

```python
# Biblioteca coerenciaBERT
from coerenciabert import CoerenciaBERT

# Instancia um objeto da classe CoerenciaBERT e recupera o MCL BERT especificado

# coerenciabert = CoerenciaBERT("neuralmind/bert-base-portuguese-cased") # BERTimbau base
coerenciabert = CoerenciaBERT("neuralmind/bert-large-portuguese-cased") # BERTimbau large
# coerenciabert = CoerenciaBERT("bert-base-multilingual-cased") # BERT Multilingual
```

## Exemplo de texto 1

Medindo documento com sentenças.

```python
# Documento e suas sentenças
DO = ["Bom Dia, professor.",
      "Qual o conteúdo da prova?",
      "Vai cair tudo na prova?",
      "Aguardo uma resposta, João."]

# Recupera a medida Ccos do documento
CcosDO = coerenciabert.getMedidaCoerenciaCosseno(DO)

# Mostra a medida recuperada
print("Ccos DO    :", CcosDO) #Ccos DO    : 0.8178287347157797
```

## Exemplo de texto 2

Comparando documento original(DO) com documento com sentenças permutadas(permDO).

```python
# Define documento original com 4 sentenças
DO1 = ["Bom Dia, professor.",
      "Qual o conteúdo da prova?",
      "Vai cair tudo na prova?",
      "Aguardo uma resposta, João."]      

# Define um documento com a permutação das sentenças do documento original
permDO1 = ["Aguardo uma resposta, João.",
           "Qual o conteúdo da prova?",              
           "Bom Dia, professor.",
           "Vai cair tudo na prova?"]    

# Recupera a medida Ccos dos documentos
CcosDO1 =  coerenciabert.getMedidaCoerenciaCosseno(DO1)
CcospermDO1 =  coerenciabert.getMedidaCoerenciaCosseno(permDO1)

# Mostra as medidas recuperadas
print("Ccos DO    :", CcosDO1)
print("Ccos permDO:", CcospermDO1)
```

## Exemplo de texto 3

Comparando documento original(DO) com documento com sentenças permutadas(permDO).

```python
# Define documento original com 3 sentenças
DO2 = ["Hoje acordei cedo.",
      "Tomei o meu café da manhã e fui correr."
      "Na volta da corrida tomei um banho e fui trabalhar."]

# Define um documento com a permutação das sentenças do documento original
permDO2 = ["Hoje não acordei tarde.", 
          "Comi o meu café da manhã correndo"
          "Na ida da corrida me sujei e fui trabalhar."]               

# Recupera a medida Ccos do documento
CcosDO2 = coerenciabert.getMedidaCoerenciaCosseno(DO2)
CcospermDO2 =  coerenciabert.getMedidaCoerenciaCosseno(permDO2)

# Mostra as medidas recuperadas
print("Ccos DO2    :", CcosDO2)
print("Ccos permDO2:", CcospermDO2)
```
